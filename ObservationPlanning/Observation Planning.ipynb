```python
import astroplan as ap
import numpy as np
import matplotlib.pyplot as plt 
from astroplan import Observer
from astropy import units as u
from astropy import constants as c
from astropy.time import Time
from astropy.coordinates import solar_system_ephemeris, EarthLocation
from astropy.coordinates import get_body_barycentric, get_body, get_moon 
from astropy.coordinates import SkyCoord
from astroplan import download_IERS_A
from astroplan import FixedTarget
download_IERS_A()
```

```python
philly_loc = EarthLocation(lat='39d57m35s', lon='75d11m49s')
philly_obs = ap.Observer(location = philly_loc, name = 'Philadelphia', timezone = 'US/Eastern')
```

```python
altair = ap.FixedTarget.from_name('Altair')
vega = ap.FixedTarget.from_name('Vega')
m82 = ap.FixedTarget.from_name('m82')
capella_coords = SkyCoord('5h16m41.6s', '45d59m44.7s', frame='icrs')
capella = ap.FixedTarget(name='Capella', coord=capella_coords)
```

```python
time_obs = Time('2018-02-01 12:00:00') #UTC
transit = philly_obs.target_meridian_transit_time(time_obs, capella)
print(transit.datetime)
```

```python
from astroplan.plots import plot_airmass, plot_sky
plot_airmass(m82, philly_obs, time_obs)
plt.show()
```

```python
plt.figure(figsize=(10, 7))
t = time_obs+np.arange(-4, 5)*u.hr
plot_sky(m82, philly_obs, time_obs)
plt.show()
```
